```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(-2, 2, 1000)
y = np.linspace(-2, 2, 1000)
xc, yc = np.meshgrid(x, y)
rc = np.sqrt(xc**2 + yc**2)
zc = np.sin(np.pi * rc)/(np.pi * rc)
```

```python
plt.rcParams["figure.figsize"] = (20, 20)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(xc, yc, zc, cmap='ocean')
plt.show()
```

```python
zc = 2*rc**2 * np.exp(-rc**2)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(xc, yc, zc, cmap='ocean', alpha=0.9)
plt.show()
```

```python
zc = xc * yc 
```

```python
ax = plt.axes()
ax.contour(xc, yc, zc)
plt.show()
```

```python
zc = xc**3 + xc**2 - yc**2
```

```python
x = np.linspace(1, 50, 1000)
y = np.linspace(1, 50, 1000)
xc, yc = np.meshgrid(x, y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xc, yc, zc, cmap='ocean', alpha=0.9)

angle=30
ax.view_init(30, angle)
plt.show()
```
