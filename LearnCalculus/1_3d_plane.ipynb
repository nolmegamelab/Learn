```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)
x_c, y_c = np.meshgrid(x, y)
z_c = x_c * 3 + y_c * 2 + 1
z_c[1:5]
```

```python
plt.rcParams["figure.figsize"] = (10,10)

ax.view_init(160, 30)
ax = plt.axes(projection='3d')
ax.plot_surface(x_c, y_c, z_c, cmap="ocean")
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x_c, y_c = np.meshgrid(x, y)
z_c = x_c + 2 * y_c + 3
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(x_c, y_c, z_c, cmap='ocean')
plt.show()
```

```python
ax = plt.axes()
ax.contour(x_c, y_c, z_c)
plt.show()
```

정리 2.3.1 directional derivative의 일차함수 버전.

```python
x = np.linspace(-20, 20, 100)
y = np.linspace(-20, 20, 100)
xc, yc = np.meshgrid(x, y)
zc = xc ** 2 + yc ** 2
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(xc, yc, zc, cmap="ocean")
plt.show()
```

```python
ax = plt.axes()
ax.contour(xc, yc, zc)
plt.show()
```

```python
zc = xc**2 - yc**2
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(xc, yc, zc, cmap='ocean')
plt.show()
```

```python
ax = plt.axes()
ax.contour(xc, yc, zc)
plt.show()
```

```python
zc = xc**2 / 9 + yc**2/4
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(xc, yc, zc, cmap='ocean')
plt.show()
```

```python
levels = np.linspace(1, 100, 20)
ax = plt.axes()
ax.contour(xc, yc, zc, levels=levels)
plt.show()
```
